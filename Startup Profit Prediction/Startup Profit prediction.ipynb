### Model Setting

#### Splitting dataset into train and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
reg_model=linear_model.LinearRegression()
reg_model=LinearRegression().fit(x_train,y_train)
```

```python
print(reg_model.intercept_)
print(reg_model.coef_)
```

#### Predicting the Test and Train set result


```python
y_pred = reg_model.predict(x_test)
x_pred = reg_model.predict(x_train)
```

#### Actual value and the predicted value of test set

```python
reg_model_diff=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred })
reg_model_diff
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = np.sqrt(metrics.mean_squared_error(y_test, y_pred))

print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', r2)
```

#### Actual value and the predicted value of train set

```python
reg_model_diff=pd.DataFrame({'Actual Value':y_train,'Predicted Value':x_pred })
reg_model_diff
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

mae = metrics.mean_absolute_error(y_train, x_pred)
mse = metrics.mean_squared_error(y_train, x_pred)
r2 = np.sqrt(metrics.mean_squared_error(y_train, x_pred))

print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', r2)
```
